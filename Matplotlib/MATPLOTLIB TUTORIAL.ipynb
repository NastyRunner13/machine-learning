```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [2,4,6,8,10]

plt.figure(figsize=(5,3), dpi=100)

plt.plot(x, y, label="2x", color="red", linewidth=0, marker='x', markersize=10)

x2 = np.arange(0,5,0.5)
print(x2)
plt.plot(x2[:6], x2[:6]**2, 'r', label="x^2")
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.title("Our first graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

plt.legend()

# plt.savefig('mygraph.png', dpi=300)

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']

print(patterns)

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))

plt.show()
```

### Real World Examples

#### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(12,8))

plt.title('Gas Prices Over Time in USD', fontdict={'fontweight': 'bold', 'fontsize':18})

for country in gas:
    if country != "Year":
        plt.plot(gas.Year, gas[country], marker="." ,label="USA")



# plt.plot(gas.Year, gas.USA, "r.-" ,label="USA")
# plt.plot(gas.Year, gas.Canada,"b.-", label="Canada")
# plt.plot(gas.Year, gas["South Korea"],"c.-", label="South Korea")
# plt.plot(gas.Year, gas.Australia,"y.-", label="Australia")

plt.xticks(gas.Year[::3])

plt.xlabel("Year")
plt.ylabel("USD")

plt.legend()

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

#### Histograms

```python
bins = []

for i in range(40,101):
    if i==0 or i%10==0:
        bins.append(i)

plt.figure(figsize=(10, 8))

plt.title('Distribution of Player Skills in FIFA 2018')

plt.hist(fifa.Overall, bins=bins)

plt.xlabel('Skill level')
plt.ylabel('Number of Players')

plt.xticks(bins)

plt.show()
```

#### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left' , 'Right']
colors = ['blue', 'red']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Prefrence of FIFA Players', fontdict={'color':'white'})

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150) ].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 170) ].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 2000) ].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200) ].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (1, 0 ,0, 0 ,0)

plt.style.use('ggplot')

plt.figure(figsize=(10,10))
plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.show()
```

```python
plt.style.use('default')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revls = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

boxes = plt.boxplot([barcelona, madrid, revls], labels=labels, patch_artist=True, medianprops={'linewidth': 2})

for box in boxes['boxes']:
    box.set(color="#4286f4", facecolor='#e0e0e0' ,linewidth=2)

plt.title("Professional Soccer Team Comaprison")
plt.ylabel('FIFA Overall Rating')

plt.show()
```
